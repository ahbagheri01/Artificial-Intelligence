            <br/>
            در تابع دوم باید الگوریتم gradient descent
            را پیاده‌سازی کنید.
            در ورودی تابع، خود تابع f،
            مقادیر اولیه،

            و همچنین تعداد مراحل اجرای الگوریتم داده شده است.

            نیز ورودی داده‌شده است و هرگاه

            بیشتر شد، اجرای الگوریتم را قطع کنید و خروجی‌ها را تا همان لحظه خروجی دهید.
            درنهایت تابع باید
            دو آرایه خروجی دهد که آرایه‌ی اول مقادیر x
            در طول زمان و آرایه‌ی دوم مقادیر y در طول زمان است.
            یعنی مثلا 
            x_points[i]
            باید نشان دهنده‌ی مقدار x

```python
def gradient_update(f, x0, y0, alpha):
    epsilon = 10e-6
    new_x = x0 - alpha * ((f(x0+epsilon,y0) - f(x0,y0))/epsilon)
    new_y = y0 - alpha * ((f(x0,y0+epsilon) - f(x0,y0))/epsilon)
    #     todo
    return new_x, new_y
    
def gd(f, initial, alpha, thereshold, steps):
    x ,y = initial[0],initial[1]
    x_points, y_points = [x],[y]
    count = 0
    while count < steps and abs(x) <= thereshold and abs(y) <= thereshold:
        x,y = gradient_update(f,x,y,alpha) 
        x_points.append(x)
        y_points.append(y)
        count+=1
    return x_points, y_points
```

```python
def f(x, y):
    return x**2 + y**2 + x*y
```

```python
v_func = np.vectorize(f)   

x, y = np.meshgrid(np.linspace(-50, 50, 100),
                   np.linspace(-50, 50, 100))

alpha_1, alpha_2,alpha_3 = 0.05, 0.25, 0.8
xs_1, ys_1 = gd(f, [-50, 50], alpha_1, 50, 1000)
xs_2, ys_2 = gd(f, [-50, 50], alpha_2, 50, 1000)
xs_3, ys_3 = gd(f, [-50, 50], alpha_3, 50, 1000)
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,5))
ax1.contour(x, y, v_func(x, y))
ax2.contour(x, y, v_func(x, y))
ax3.contour(x, y, v_func(x, y))
ax1.scatter(xs_1, ys_1, s=2, c="r")
ax2.scatter(xs_2, ys_2, s=2, c="g")
ax3.scatter(xs_3, ys_3, s=2, c="b")
print("report alpha:"+str(alpha_1))
print("steps: "+str(len(xs_1) -1))
print("min converged: "+str(f(xs_1[len(xs_1) -1] , ys_1[len(ys_1)-1])))
print("report alpha:"+str(alpha_2))
print("steps: "+str(len(xs_2) -1))
print("min converged: "+str(f(xs_2[len(xs_2)-1 ] , ys_1[len(ys_2)-1 ])))
print("report alpha:"+str(alpha_3))
print("steps: "+str(len(xs_3) -1))
print("min converged: "+"undesirable divergent")
plt.show()
```

            ب)
            تابع زیر را کامل کنید.
            این تابع
            مقدار لحظه‌ای بردار w
            مجموعه‌ی داده‌ی U
            و همچنین Z
            را ورودی می‌گیرید.
            Z 
            شامل آرایه‌ای از برچسب‌های داده‌ها است.
            U
            نیز به صورت آرایه‌ای به فرم
            $U[i] = (x_i, y_i, 1)$
            ورودی داده می‌شود.
            همچنین G نیز در ورودی تابع داده می‌شود.

```python
def compute_loss(w, G, U, Z,N):
    loss = 1/2 * np.dot(w,w) 
    temp = 0
    for i in range(N):
        temp+=max(0,1-Z[i]*np.dot(w,U[i]))
    temp *= G/N
    loss+=temp
    return loss
```

```python
def compute_gradient(w, G, u, z):
    return (w if max(0,1-z*(np.dot(w,u)) ) == 0 else w-np.multiply(G*z,u))
```

```python
import tqdm as tqdm
def GD(U, Z,N,STEPS = 10000):
    G = 10000
    LEARNING_STEP = 0.000001
    losses = []
    w = [0.0, 0.0, 0.0]
    for i in tqdm.tqdm(range(STEPS)):
        gradient = np.array([0.0, 0.0, 0.0])
        for j in range(N):
            gradient += compute_gradient(w, G, U[j], Z[j])
        losses.append(compute_loss(w, G, U, Z,N))
        w -= (LEARNING_STEP / N) * gradient
    print("CALCULATED WEIGHTS: ", w)
    return w, losses
U = []
x  = []
N = len(train_dataset.x)
for i in range(N):
    U.append([train_dataset.x[i],train_dataset.y[i],1])
    x.append(train_dataset.x[i])
    

Z = train_dataset.z
W, L = GD(U, Z,N)
```
